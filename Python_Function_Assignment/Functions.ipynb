1. **What is the difference between a function and a method in Python?**

    **Function**
    * A function is a block of reusable code defined using the **def** keyword.
    * It is independent and not tied to any object or class (unless you explicitly put it inside a class).
    * You call it by its name directly.
    ```
    # A simple function
    def greet(name):
        return f"Hello, {name}!"

    print(greet("Shivam"))   # Output: Hello, Shivam!
    ```

  **Method**
    * A method is a function that is associated with an object/class.
    * It is defined inside a class and is called using the dot (.) notation on an object.
    * It automatically takes the instance (self) as its first argument.

    ```
    class Greeter:
    def greet(self, name):   # 'self' refers to the object calling the method
        return f"Hello, {name}!"

    # Create an object of the class
    g = Greeter()
    print(g.greet("Shivam"))   # Output: Hello, Shivam!
    ```

2. **Explain the concept of function arguments and parameters in Python.**

    **Parameters** are the variables listed inside the parentheses in the function definition.

    **Arguments** are the actual values passed to the function when calling it.

    Python supports different types of arguments: positional, keyword, default, and variable-length.

    Proper matching of arguments with parameters is crucial for the function to work correctly.
    
    ```
    # Here 'a' and 'b' are PARAMETERS
    def add(a, b):
        return a + b

    # Here 5 and 10 are ARGUMENTS
    result = add(5, 10)
    print("Result:", result)   # Output: Result: 15
    ```

3. **What are the different ways to define and call a function in Python?**

    In Python, we can define functions in multiple ways. Let’s cover them below with examples for each.

    * Standard Function Definition :- Using the def keyword.
    ```
    def greet(name):
    return f"Hello, {name}!"

    print(greet("Shivam"))   # Calling the function
    ```

    * Function with Default Arguments :- You can provide default values for parameters.
    ```
    def greet(name="Guest"):
    return f"Hello, {name}!"

    print(greet())           # Hello, Guest
    print(greet("Shivam"))   # Hello, Shivam
    ```

    * Function with Variable-length Arguments :-
      * *args :- accepts multiple positional arguments.
      * **kwargs :- accepts multiple keyword arguments.

      ```
      def show(*args, **kwargs):
          print("Args:", args)
          print("Kwargs:", kwargs)

      show(1, 2, 3, name="Shivam", age=23)
      # Args: (1, 2, 3)
      # Kwargs: {'name': 'Shivam', 'age': 23}
      ```

    * Lambda (Anonymous) Function :- Defined using lambda keyword (single-line functions).
    ```
    square = lambda x: x * x
    print(square(5))   # 25
    ```

    * Nested Function :- Defining a function inside another function.
    ```
    def outer():
        def inner():
            return "I am inner function"
        return inner()

    print(outer())   # I am inner function
    ```

5. **What are iterators in Python and how do they differ from iterables?**

    **Iterable**

    An iterable is any Python object that can return its elements one by one.

    Examples: list, tuple, string, dictionary, set.

    They can be converted into an iterator using the iter() function.

    ```
    my_list = [1, 2, 3]

    # This is an iterable (list)
    for item in my_list:
        print(item)   # 1, 2, 3
    ```

    **Iterator**

    An iterator is an object that produces items of an iterable one at a time using the __next__() method.

    You can get an iterator from an iterable using the built-in iter() function.

    To get the next value, you use next().

    Iterators remember their state (they don’t restart unless re-created).

    ```
    my_list = [1, 2, 3]

    # Convert iterable into an iterator
    it = iter(my_list)

    print(next(it))   # 1
    print(next(it))   # 2
    print(next(it))   # 3
    # print(next(it))   # Raises StopIteration
    ```



6. **Explain the concept of generators in Python and how they are defined.**

    A generator is a special type of iterator in Python.

    It is used to generate values one at a time using the yield keyword instead of return.

    Generators are memory-efficient because they don’t store all values in memory at once — they produce them on demand (lazy evaluation).

    **Define Generator function**

    Defined like a normal function, but uses yield instead of return.

    Each time yield is hit, the function "pauses" and resumes from there on the next call.

    ```
    def countdown(n):
        while n > 0:
            yield n
            n -= 1

    # Create generator
    gen = countdown(3)

    print(next(gen))   # 3
    print(next(gen))   # 2
    print(next(gen))   # 1
    # print(next(gen)) # Raises StopIteration
    ```

7. **What are the advantages of using generators over regular functions?**

    * Memory Efficient: Generators don’t store all values in memory; they yield one at a time.

    * Lazy Evaluation: Values are produced only when requested, not all at once.

    * Improved Performance: Suitable for large datasets or streams.

    * Simplifies Code: Cleaner syntax for iterators using yield.

    ```
    def natural_numbers():
    n = 1
    while True:
        yield n
        n += 1

    gen = natural_numbers()
    print(next(gen))  # 1
    print(next(gen))  # 2
    print(next(gen))  # 3
    ```


8. **What is a lambda function in Python and when is it typically used?**

    A lambda function is a small anonymous function in Python.

    Defined using the keyword lambda instead of def.

    Can take any number of arguments but must contain only one expression.

    Automatically returns the result of the expression (no need to write return).

    ```
    # Normal function
    def square(x):
        return x * x

    # Lambda equivalent
    square_lambda = lambda x: x * x

    print(square_lambda(5))   # 25
    ```

    **Lambda Usecase**
    
    Short, throwaway functions (when you don’t want to define a full def function).

    Inside built-in functions like map(), filter(), and sorted().

    For one-liners where defining a full function would be unnecessary.

    ```
    nums = [1, 2, 3, 4]
    squares = list(map(lambda x: x*x, nums))
    print(squares)   # [1, 4, 9, 16]
    ```

9. **Explain the purpose and usage of the `map()` function in Python.**

    The map() function in Python is used to apply a function to every item of an iterable (like list, tuple, etc.) and return an iterator with the results.

    Useful for transforming data without using explicit loops.

    ```
    def square(x):
        return x * x

    numbers = [1, 2, 3, 4]
    result = map(square, numbers)

    print(list(result))   # [1, 4, 9, 16]
    ```

10. **What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?**

    **map()**

    Purpose :- Transforms each element of an iterable using a function.

    Returns :- An iterator with the transformed values.

    ```
    numbers = [1, 2, 3, 4]
    squares = map(lambda x: x*x, numbers)
    print(list(squares))   # [1, 4, 9, 16]
    ```

    **filter()**

    Purpose :- Filters elements of an iterable based on a condition (True/False).

    Returns :- An iterator containing only the elements that passed the condition.

    ```
    numbers = [1, 2, 3, 4, 5, 6]
    evens = filter(lambda x: x % 2 == 0, numbers)
    print(list(evens))   # [2, 4, 6]
    ```

    **reduce()**

    Purpose :- Reduces an iterable to a single cumulative value by applying a function repeatedly.

    Returns :- A single value.

    Comes from functools module (not built-in).

    ```
    from functools import reduce

    numbers = [1, 2, 3, 4]
    product = reduce(lambda x, y: x * y, numbers)
    print(product)   # 24  (1*2*3*4)
    ```

````python
# 1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.
num_list = [1, 2, 3, 4, 5, 6]
def sumOfEvenNumber(num_list):
    sum = 0
    for i in num_list:
        if i % 2 == 0:
            sum = sum + i
    return sum
print(sumOfEvenNumber(num_list))
````

````python
#5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
class Fibonacci:
    def __init__(self, x):
        self.x = x
        self.count = 0
        self.a = 0
        self.b = 1

    def __iter__(self):
        return self  # return the iterator object

    def __next__(self):
        if self.count >= self.x:
            raise StopIteration  # stop when limit is reached

        if self.count == 0:
            self.count += 1
            return self.a
        elif self.count == 1:
            self.count += 1
            return self.b
        else:
            z = self.a + self.b  # next term is sum of previous two
            self.a, self.b = self.b, z  # update values
            self.count += 1
            return z
n = int(input("Enter number"))
fib = Fibonacci(n)

print("Fibonacci seq")
for num in fib:
    print(num)
````

````python
#6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
def powerOfTwoExp(max_exp):
    for i in range(max_exp + 1):  # loop from 0 to max_exp
        yield 2 ** i
for num in powerOfTwoExp(10):
    print(num)
````

````python
#10. Create a Python program that uses `filter()` to remove all the vowels from a given string
def removeAllVowelFromString(s):
    stri = 'aeiouAEIOU'
    return ''.join(filter(lambda char: char not in stri, s))
text = "Hi my name is Shivam Singh"
result = removeAllVowelFromString(text)
print("vowels removed string: ", result)
````
