1. **What is a random variable in probability theory?**

    A **random variable** is a function that assigns a real number to each possible outcome of a **random experiment**.

    In simple terms, it’s a way to **numerically describe** the outcome of a random process.

    Formally, if the sample space of a random experiment is denoted by **S**, then a random variable **X** is a function:

    $$
    X: S \rightarrow \mathbb{R}
    $$

    That means for every outcome $s \in S$, $X(s)$ gives a numerical value.

    For example:
   - Tossing a coin → {Heads, Tails}
   - Assigning values: Heads → 1, Tails → 0  
    → Random variable \( X \) can be defined such that:
    $$
    X(\text{Heads}) = 1, \quad X(\text{Tails}) = 0
    $$

2. **What are the types of random variables?**

   a) Discrete Random Variable
    A **discrete random variable** takes on a **countable number** of distinct values.

    Examples:
    - Number of heads when tossing 3 coins → {0, 1, 2, 3}
    - Number of customers arriving in a minute

    It is described using a **probability mass function (PMF)**:
    $$
    P(X = x_i) = p_i, \quad \text{where } \sum p_i = 1
    $$


   b) Continuous Random Variable
    A **continuous random variable** takes on **uncountably infinite** values (usually real numbers within an interval).

    Examples:
    - Height of students in a class
    - Time taken to finish a task

    It is described using a **probability density function (PDF)**:
    $$
    P(a \le X \le b) = \int_a^b f(x) \, dx
    $$
    where \( f(x) \ge 0 \) and
    $$
    \int_{-\infty}^{\infty} f(x) \, dx = 1
    $$

   **Example**

   **Experiment:** Rolling a fair six-sided die  
   Sample space:  
   $$
   S = \{1, 2, 3, 4, 5, 6\}
   $$

   Define a random variable \( X \) = “the number shown on the die”.

   | Outcome (s) | 1 | 2 | 3 | 4 | 5 | 6 |
   |--------------|---|---|---|---|---|---|
   | X(s)         | 1 | 2 | 3 | 4 | 5 | 6 |
   | P(X = x)     | 1/6 | 1/6 | 1/6 | 1/6 | 1/6 | 1/6 |

   Example in Real Life

- **Banking:** Number of customers arriving per hour → Discrete random variable  
- **Engineering:** Time before a component fails → Continuous random variable  
- **Education:** Marks obtained in an exam → Continuous random variable

3.  **Explain the difference between discrete and continuous distributions?**

    In probability theory, **distributions** describe how the values of a **random variable** are spread or distributed.  
    They can be broadly classified into **Discrete** and **Continuous** distributions.

    - **Conceptual Difference**

    | Feature | **Discrete Distribution** | **Continuous Distribution** |
    |----------|---------------------------|------------------------------|
    | **Definition** | Describes the probabilities of **discrete (countable)** outcomes. | Describes probabilities over a **continuous range** of outcomes. |
    | **Random Variable Type** | Discrete random variable | Continuous random variable |
    | **Possible Values** | Finite or countably infinite (e.g., 0, 1, 2, 3, …) | Uncountably infinite (any real number in an interval) |
    | **Probability Representation** | Probability Mass Function (**PMF**) | Probability Density Function (**PDF**) |
    | **Probability of a Single Value** | Can be **non-zero**, e.g. $ P(X=3) = 0.2 $ | Always **zero**, $ P(X = a) = 0 $; probabilities are found over intervals |
    | **Graph Type** | Bar graph (distinct values) | Smooth curve (continuous range) |
    | **Examples** | Number of heads in 3 coin tosses, number of students in a class | Height, weight, temperature, time taken to complete a task |

    ---

    **Mathematical Representation**

    **Discrete Distribution**
    For a discrete random variable \( X \):
    $$
    P(X = x_i) = p_i
    $$
    where
    $$
    \sum_i p_i = 1
    $$

    **Example:** Tossing a fair coin  
    $$
    P(X = 0) = 0.5, \quad P(X = 1) = 0.5
    $$


    **Continuous Distribution**
    For a continuous random variable \( X \):
    $$
    P(a \le X \le b) = \int_a^b f(x) \, dx
    $$
    where
    $$
    \int_{-\infty}^{\infty} f(x) \, dx = 1
    $$

    **Example:** Time taken to complete a task  
    The probability that time \( X \) lies between 5 and 10 seconds is:
    $$
    P(5 \le X \le 10) = \int_5^{10} f(x) \, dx
    $$

4. **What is a binomial distribution, and how is it used in probability?**

    A **Binomial Distribution** is a **discrete probability distribution** that describes the number of **successes** in a fixed number of **independent Bernoulli trials**, where each trial has only two possible outcomes:
    - **Success (with probability p)**
    - **Failure (with probability q = 1 - p)**

    It is denoted as:

    $$
    X \sim \text{Binomial}(n, p)
    $$

    where:
    - $ n $ = number of trials  
    - $ p $ = probability of success on each trial  
    - $ q = 1 - p $ = probability of failure  
    - $ X $ = number of successes in $ n $ trials


    **Probability Mass Function (PMF)**

    The probability of getting exactly $ k $ successes in $ n $ trials is given by:

    $$
    P(X = k) = \binom{n}{k} p^k (1 - p)^{n - k}
    $$

    where:
    $$
    \binom{n}{k} = \frac{n!}{k!(n - k)!}
    $$

    for $$ k = 0, 1, 2, \dots, n $$

    Imagine you flip a fair coin 10 times.  
    Each flip has two outcomes: **Head (success)** or **Tail (failure)**.

    - $ n = 10 $ (trials)  
    - $ p = 0.5 $ (probability of Head)  
    - $ X $ = number of Heads obtained

    Then $ X $ follows a **Binomial Distribution**, and you can use it to find the probability of getting exactly 6 heads:

    $$
    P(X = 6) = \binom{10}{6} (0.5)^6 (0.5)^4 = 210 \times (0.5)^{10} = 0.205
    $$

    So, there’s a **20.5% chance** of getting exactly 6 heads.


5. **What is the standard normal distribution, and why is it important?**

    The **Standard Normal Distribution** is a special case of the **Normal Distribution** (also called the **Gaussian Distribution**) that has:

    $$
    \mu = 0 \quad \text{(mean)} \\
    \sigma = 1 \quad \text{(standard deviation)}
    $$

    It is denoted as:

    $$
    Z \sim N(0, 1)
    $$

    The **probability density function (PDF)** of the standard normal distribution is:

    $$
    f(z) = \frac{1}{\sqrt{2\pi}} e^{-\frac{z^2}{2}}
    $$

    where $ z $ is any real number.



    **Intuitive Understanding**

    A **normal distribution** describes data that is symmetrically distributed around a mean, forming the classic **bell-shaped curve**.

    When we **standardize** a normal distribution (by converting data into **Z-scores**), we get the **standard normal distribution**, centered at zero with a standard deviation of one.

    This transformation allows us to compare different datasets on the same scale.


    **Formula for Z-Score**

    To convert a value $ X $ from any normal distribution $ N(\mu, \sigma) $ to the standard normal form:

    $$
    Z = \frac{X - \mu}{\sigma}
    $$

    where:
    - $ X $ = data point
    - $ \mu $ = mean of the original distribution
    - $ \sigma $ = standard deviation of the original distribution
    - $ Z $ = standardized value (Z-score)

    **Example**

    Suppose the heights of students are normally distributed with:

    $$
    \mu = 170 \text{ cm}, \quad \sigma = 10 \text{ cm}
    $$

    Find the Z-score for a student who is **185 cm** tall:

    $$
    Z = \frac{185 - 170}{10} = 1.5
    $$

    This means the student’s height is **1.5 standard deviations above the mean**.

    **Importance of the Standard Normal Distribution**

    * **Simplifies Probability Calculations**
        Instead of calculating probabilities for every unique normal distribution, we use **Z-tables** (or standard normal tables) to find areas under the curve for standardized Z-scores.

    * **Used for Statistical Inference**
        It is fundamental in:
        - Hypothesis testing
        - Confidence intervals
        - Regression analysis
        - Quality control

    * **Comparison Across Different Scales**
        Standardization allows comparing scores from different normal distributions (e.g., test scores, heights, weights) on the same scale.

    * **Foundation for Many Models**
        The standard normal distribution is the basis for:
      - **Z-test**
      - **T-distribution** (when σ is unknown)
      - **Chi-square** and **F-distributions**

6. **What is the Central Limit Theorem (CLT), and why is it critical in statistics?**

    The **Central Limit Theorem (CLT)** is one of the most important concepts in statistics and probability theory.  
    It states that:

    > **When you take a large number of random samples from any population (with finite mean and variance),  
    > the distribution of the sample means will approach a normal distribution —  
    > regardless of the shape of the original population.**

    ---

    Mathematical Formulation

    Let:
    - $ X_1, X_2, X_3, \dots, X_n $ be independent and identically distributed (i.i.d.) random variables  
    - Each with mean $ \mu $ and standard deviation $ \sigma $

    Then, the **sample mean** is:

    $$
    \bar{X} = \frac{1}{n} \sum_{i=1}^{n} X_i
    $$

    According to the **Central Limit Theorem**:

    $$
    Z = \frac{\bar{X} - \mu}{\sigma / \sqrt{n}} \rightarrow N(0, 1) \text{ as } n \rightarrow \infty
    $$

    That means the standardized sample mean follows a **standard normal distribution** when the sample size \( n \) is sufficiently large.

    **Example: Rolling a Die**

    - Population (1 roll): Uniform distribution → {1, 2, 3, 4, 5, 6}
    - Take multiple samples:
      - Mean of 2 rolls → distribution starts to look smoother
      - Mean of 10 rolls → bell-shaped curve appears
      - Mean of 30 rolls → curve closely resembles a normal distribution

    As **sample size increases**, the shape of the distribution of means becomes **more normal**.

    **Why the CLT is Critical in Statistics**

    | Reason | Explanation |
    |---------|--------------|
    | **1. Enables Use of Normal Distribution** | Lets us apply normal distribution techniques to problems involving means, even if the population is not normal. |
    | **2. Foundation for Inferential Statistics** | Underpins hypothesis testing, confidence intervals, and regression analysis. |
    | **3. Simplifies Analysis** | Makes it possible to use Z-scores and probability tables for complex real-world data. |
    | **4. Real-World Application** | Almost all sample-based statistics (averages, proportions, etc.) rely on the CLT to justify normal approximations. |
    | **5. Practical Universality** | Explains why many natural phenomena (like IQ, measurement errors, heights, etc.) follow a bell curve. |


7. What is the significance of confidence intervals in statistical analysis?

    A **Confidence Interval (CI)** is a range of values, derived from sample data, that is **likely to contain the true population parameter** (such as mean or proportion) with a certain level of confidence.

    In simple terms:
    > A confidence interval gives an estimated range for a population value (like the true mean), based on the data from a sample.

    ---

    **General Formula**

    $$
    \text{Confidence Interval} = \text{Point Estimate} \; \pm \; (\text{Critical Value} \times \text{Standard Error})
    $$

    where:
    - **Point Estimate** = sample statistic (e.g., sample mean $ \bar{X} $)
    - **Critical Value** = depends on the desired confidence level (e.g., 1.96 for 95% confidence under normal distribution)
    - **Standard Error (SE)** = variability of the statistic  
      $$
      SE = \frac{\sigma}{\sqrt{n}} \text{ (for mean)}
      $$

    **Why Confidence Intervals Are Significant**

    | Significance | Explanation |
    |---------------|--------------|
    | **1. Quantifies Uncertainty** | Provides a range, not just a single estimate, showing how precise the sample estimate is. |
    | **2. Improves Decision-Making** | Helps decide if sample results are statistically meaningful. |
    | **3. Basis for Hypothesis Testing** | If a hypothesized population value (like mean = 0) lies outside the CI, it suggests statistical significance. |
    | **4. Enables Comparison** | Overlapping confidence intervals can indicate non-significant differences between groups. |
    | **5. Widely Used in Research** | Used in medicine, economics, machine learning, and social sciences to express result reliability. |


8. **What is the concept of expected value in a probability distribution?**

    The **expected value (EV)** of a random variable is the **long-run average** value it takes after many repetitions of an experiment.  
    It represents the **theoretical mean** of a probability distribution — the value you’d “expect” on average if the process were repeated infinitely many times.

    Mathematically, it provides a measure of **central tendency** for random variables.

    **Mathematical Formulas**

    For a Discrete Random Variable

    If a discrete random variable $ X $ takes values $ x_1, x_2, ..., x_n $ with corresponding probabilities $ P(x_1), P(x_2), ..., P(x_n) $, then:

    $$
    E(X) = \sum_{i=1}^{n} x_i \cdot P(x_i)
    $$

    For a Continuous Random Variable

    If a continuous random variable $ X $ has a probability density function (PDF) $ f(x) $, then:

    $$
    E(X) = \int_{-\infty}^{\infty} x \cdot f(x) \, dx
    $$


    Example (Discrete Case)

    Suppose a fair six-sided die is rolled.  
    Possible outcomes: $ X = 1, 2, 3, 4, 5, 6 $  
    Each with probability $ P(X) = \frac{1}{6} $.

    $$
    E(X) = \sum_{i=1}^{6} x_i \cdot P(x_i)
    $$

    $$
    E(X) = (1 + 2 + 3 + 4 + 5 + 6) \times \frac{1}{6} = \frac{21}{6} = 3.5
    $$

    The **expected value** is **3.5**, meaning the average of many rolls tends to approach 3.5.


    Example (Continuous Case)

    Let $ X $ be a continuous random variable uniformly distributed on $[0, 1]$.  
    Then $ f(x) = 1 $ for $ 0 \le x \le 1 $.

    $$
    E(X) = \int_{0}^{1} x \cdot 1 \, dx = \frac{x^2}{2} \Big|_0^1 = \frac{1}{2}
    $$

    The **expected value** is **0.5**.

    | Type of Variable | Formula | Example | Result |
    |------------------|----------|----------|---------|
    | Discrete | $ E(X) = \sum x_i P(x_i) $ | Fair die | 3.5  
    | Continuous | $ E(X) = \int x f(x) dx $ | Uniform(0,1) | 0.5 |


```python
# 9. Write a Python program to generate 1000 random numbers from a normal
# distribution with mean = 50 and standard deviation = 5. Compute its mean and standard
# deviation using NumPy, and draw a histogram to visualize the distribution.

import numpy as np
import matplotlib.pyplot as plt

# Generate 1000 random numbers from a normal distribution
data = np.random.normal(loc=50, scale=5, size=1000)

# Compute mean and standard deviation
mean = np.mean(data)
std_dev = np.std(data)

# Display results
print(f"Mean of the distribution: {mean:.2f}")
print(f"Standard Deviation of the distribution: {std_dev:.2f}")

# Draw a histogram
plt.figure(figsize=(8, 5))
plt.hist(data, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
plt.title("Histogram of Normally Distributed Random Numbers")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

9. **You are working as a data analyst for a retail company. The company has
collected daily sales data for 2 years and wants you to identify the overall sales trend.
daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
 235, 260, 245, 250, 225, 270, 265, 255, 250, 260]
● Explain how you would apply the Central Limit Theorem to estimate the average sales
with a 95% confidence interval.
● Write the Python code to compute the mean sales and its confidence interval?.**

    The **Central Limit Theorem (CLT)** states that if we take repeated random samples from any population and compute their means,  
    the **sampling distribution of those means** will tend to follow a **normal distribution**,  
    **regardless of the shape of the original population**, as long as the sample size is sufficiently large (n ≥ 30 is common).

    Even with smaller samples, the CLT allows us to make inferences about the population mean using the **t-distribution**  
    if the population standard deviation is unknown.

    We have a sample of **daily sales data** for 20 days.  
    We want to estimate the **average daily sales** for the population with a **95% confidence interval**.

```python
import numpy as np
import scipy.stats as stats

# Daily sales data (sample)
daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
               235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

# Convert to numpy array
data = np.array(daily_sales)

# Compute sample mean and standard deviation
mean_sales = np.mean(data)
std_sales = np.std(data, ddof=1)  # ddof=1 for sample standard deviation
n = len(data)

# Compute the standard error
standard_error = std_sales / np.sqrt(n)

# 95% Confidence Interval using t-distribution
confidence = 0.95
t_critical = stats.t.ppf((1 + confidence) / 2, df=n-1)

margin_of_error = t_critical * standard_error
confidence_interval = (mean_sales - margin_of_error, mean_sales + margin_of_error)

# Display results
print(f"Sample Mean: {mean_sales:.2f}")
print(f"Sample Standard Deviation: {std_sales:.2f}")
print(f"Standard Error: {standard_error:.2f}")
print(f"t-Critical (95% CI, df={n-1}): {t_critical:.3f}")
print(f"95% Confidence Interval for Mean Sales: {confidence_interval}")
```
