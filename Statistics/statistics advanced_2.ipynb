1. **What is hypothesis testing in statistics?**

    **Hypothesis testing** is a statistical method used to make decisions or draw conclusions about a population  
    based on **sample data**.  
    It helps determine whether the observed results are **due to random chance** or **reflect a real effect**.

    In simple terms, it’s a way to **test an assumption (hypothesis)** about a population parameter.


    Key Concepts

    1. Null Hypothesis (H₀)
        The **null hypothesis** represents the default or “no effect” assumption.  
        It states that there is **no difference** or **no relationship** in the population.

        Example:  
        > H₀: The average daily sales = ₹250


    2. Alternative Hypothesis (H₁ or Ha)
        The **alternative hypothesis** is what you want to test for.  
        It represents the idea that there **is** an effect or a difference.

        Example:  
        > H₁: The average daily sales ≠ ₹250


    3. Test Statistic
        A **test statistic** is a numerical value calculated from the sample data  
        that helps decide whether to reject the null hypothesis.

        Common examples:
        - **Z-test** → for large samples or known population standard deviation.  
        - **t-test** → for small samples or unknown population standard deviation.

    4. p-value
        The **p-value** represents the probability of obtaining the observed results  
        (or more extreme ones) **if the null hypothesis were true**.

        - If **p-value ≤ α** (significance level), **reject H₀**.  
        - If **p-value > α**, **fail to reject H₀**.

        Typical α (alpha) value = **0.05** → 5% significance level.

    5. Decision Rule
   
        | Condition | Decision | Interpretation |
        |------------|-----------|----------------|
        | p ≤ 0.05 | Reject H₀ | There is significant evidence for H₁ |
        | p > 0.05 | Fail to reject H₀ | No significant evidence for H₁ |


2. **What is the null hypothesis, and how does it differ from the alternative hypothesis?**

    In **hypothesis testing**, we make assumptions about a **population parameter** (like mean or proportion)  
    and test them using **sample data**.  
    These assumptions are represented by two competing statements:

    - The **Null Hypothesis (H₀)**  
    - The **Alternative Hypothesis (H₁ or Ha)**

    Together, they form the foundation of every statistical test.

    **Null Hypothesis (H₀)**

    The **null hypothesis** represents the **default assumption** — that there is **no effect**, **no difference**,  
    or **no relationship** between variables.

    It’s the statement you **start with** and try to find evidence **against**, not for.

    Examples:
    - H₀: The average daily sales = ₹250  
    - H₀: There is no difference in test scores between two groups  
    - H₀: The new drug has the same effect as the old drug

    **Alternative Hypothesis (H₁ or Ha)**

    The **alternative hypothesis** is what you **want to test or prove**.  
    It represents the belief that there **is an effect**, **a difference**, or **a relationship**.

    Examples:
    - H₁: The average daily sales ≠ ₹250  
    - H₁: There is a difference in test scores between two groups  
    - H₁: The new drug works better than the old drug

    **Key Difference Between H₀ and H₁**

    | Feature | Null Hypothesis (H₀) | Alternative Hypothesis (H₁) |
    |----------|-----------------------|-----------------------------|
    | **Meaning** | Assumes no effect or difference | Assumes an effect or difference exists |
    | **Goal** | To be tested (and possibly rejected) | To be supported if H₀ is rejected |
    | **Symbol** | H₀ | H₁ or Ha |
    | **Example (mean)** | μ = 250 | μ ≠ 250 |
    | **Decision** | Fail to reject or reject | Supported when H₀ is rejected |
    | **Type of Test** | Two-tailed or one-tailed (depends on H₁) | Defines direction of test (≠, >, <) |

    **Example**

    Let’s say a company claims the mean weight of a product is **100 grams**.  
    You suspect it might be different.

    - **H₀:** μ = 100  
    - **H₁:** μ ≠ 100  

    After testing:
    - If **p > 0.05** → Fail to reject H₀ → No evidence that the weight differs from 100g  
    - If **p ≤ 0.05** → Reject H₀ → Evidence suggests the weight is significantly different from 100g

3. **Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.**

    The **significance level**, denoted by **α (alpha)**, is the **threshold probability** used to decide  
    whether the **null hypothesis (H₀)** should be **rejected**.

    It represents the **maximum probability** of making a **Type I error** — that is,  
    **rejecting the null hypothesis when it is actually true**.

    In simple words:  
    > The significance level tells us how much evidence we require before we say that an effect is “statistically significant.”

    **Common Values of α**

    | α (Significance Level) | Confidence Level | Typical Usage |
    |--------------------------|------------------|----------------|
    | 0.10 | 90% confidence | Preliminary studies |
    | 0.05 | 95% confidence | Standard in most research |
    | 0.01 | 99% confidence | Highly critical studies (e.g., medicine, aerospace) |

    **Role in Hypothesis Testing**

    1. **State the hypotheses:**
       - H₀: No effect or difference  
       - H₁: Effect or difference exists

    2. **Choose α (e.g., 0.05)**

    3. **Calculate the test statistic and p-value**

    4. **Compare p-value with α:**

    | Condition | Decision | Interpretation |
    |------------|-----------|----------------|
    | p ≤ α | Reject H₀ | Evidence supports the alternative hypothesis |
    | p > α | Fail to reject H₀ | Evidence not strong enough to reject the null hypothesis |


    **Example**

    Suppose you are testing whether a new teaching method improves test scores.

    - **H₀:** The new method has no effect.  
    - **H₁:** The new method improves performance.  
    - **α = 0.05**

    After analysis, the **p-value = 0.03**.

    Since **p (0.03) < α (0.05)** → **Reject H₀**.  
    There is **significant evidence** that the new teaching method improves scores.


4. **What are Type I and Type II errors? Give examples of each.**

    When we perform **hypothesis testing**, we make decisions based on sample data. However, because samples can vary, there’s always a chance we make a **wrong decision**.

    There are two types of possible errors:

    - **Type I Error (False Positive)**
    - **Type II Error (False Negative)**

    **Type I Error (α) — “False Alarm”**

    A **Type I error** occurs when we **reject the null hypothesis (H₀)** even though it is actually **true**.

    In other words, we conclude that there **is an effect or difference** when in fact **there isn’t**.

    **Probability**
    The probability of committing a Type I error = **α (significance level)**.  
    Typical α = 0.05 → means there’s a 5% chance of making this mistake.

    **Example**
    A medical test for a disease:
    - **H₀:** The patient does *not* have the disease.  
    - **H₁:** The patient *has* the disease.

    If the test wrongly indicates that the person **has the disease** when they actually don’t,  
    that’s a **Type I error**.

    **Interpretation:** A “false positive.”

    **Type II Error (β) — “Missed Detection”**

    A **Type II error** occurs when we **fail to reject the null hypothesis (H₀)** even though it is actually **false**.

    In other words, we **miss** detecting a real effect or difference.

    **Probability**
    The probability of making a Type II error = **β**.  
    The complement (1 − β) is called the **power of the test**,  
    which measures the ability to correctly detect a true effect.

    **Example**
    Continuing the medical test analogy:
    - **H₀:** The patient does *not* have the disease.  
    - **H₁:** The patient *has* the disease.

    If the test **fails to detect** the disease when the person actually *does* have it,  
    that’s a **Type II error**.

    **Interpretation:** A “false negative.”


5. **What is the difference between a Z-test and a T-test? Explain when to use each.**

    **Z-test**
    
    Both the **Z-test** and the **T-test** are statistical methods used to determine  
    whether there is a **significant difference** between a **sample mean** and a **population mean**,  
    or between **two sample means**.

    The main difference lies in **what information you have** and **the size of your sample**.

    **Z-Test**
    A **Z-test** is used when the **population standard deviation (σ)** is **known**,  
    and the **sample size (n)** is **large** (typically **n ≥ 30**).

    It assumes the **sampling distribution** of the mean is **normally distributed**.
    $$
    Z = \frac{\bar{X} - \mu}{\frac{\sigma}{\sqrt{n}}}
    $$

    Where:
    - $ \bar{X} $ = sample mean  
    - $ \mu $ = population mean  
    - $ \sigma $ = population standard deviation  
    - $ n $ = sample size  

    **When to Use a Z-Test**
    - Population standard deviation (σ) is **known**.  
    - Sample size is **large (n ≥ 30)**.  
    - Data approximately follows a **normal distribution**.
    
    Example
    A manufacturer claims that the average life of a battery is **100 hours** with a **known σ = 5**.  
    You take a sample of 50 batteries and find the sample mean = 98.  
    You can use a **Z-test** to test the claim.


    **T-Test**

    A **T-test** is used when the **population standard deviation (σ)** is **unknown**,  
    and the **sample size (n)** is **small** (typically **n < 30**).

    It uses the **sample standard deviation (s)** as an estimate for σ,  
    and the sampling distribution follows a **t-distribution** with **(n − 1)** degrees of freedom.

    $$
    t = \frac{\bar{X} - \mu}{\frac{s}{\sqrt{n}}}
    $$

    Where:
    - $ \bar{X} $ = sample mean  
    - $ \mu $ = population mean  
    - $ s $ = sample standard deviation  
    - $ n $ = sample size  

    **When to Use a T-Test**
    - Population standard deviation (σ) is **unknown**.  
    - Sample size is **small (n < 30)**.  
    - Data is approximately **normally distributed**.  

    Example
    You want to test whether the average score of 15 students differs from 70,  
    but you don’t know the population standard deviation.  
    You’ll use a **T-test**.

```python
# 8. Write a Python script to simulate data from a normal distribution and
# calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Generate data from a normal distribution
mean = 100      # population mean
std_dev = 15    # population standard deviation
n = 50          # sample size

data = np.random.normal(mean, std_dev, n)

# Sample statistics
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # sample standard deviation

# Confidence level
confidence = 0.95
alpha = 1 - confidence

# t-critical value (two-tailed)
t_critical = stats.t.ppf(1 - alpha/2, df=n-1)

# Standard error
standard_error = sample_std / np.sqrt(n)

# Margin of error
margin_of_error = t_critical * standard_error

# Confidence interval
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

# Print results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Sample Standard Deviation: {sample_std:.2f}")
print(f"t-Critical (95% CI, df={n-1}): {t_critical:.3f}")
print(f"95% Confidence Interval for the Mean: {confidence_interval}")

# Plot histogram
plt.figure(figsize=(8,5))
plt.hist(data, bins=10, color='skyblue', edgecolor='black', density=True)
plt.axvline(sample_mean, color='red', linestyle='--', label=f'Mean = {sample_mean:.2f}')
plt.axvline(confidence_interval[0], color='green', linestyle='--', label='95% CI Lower Bound')
plt.axvline(confidence_interval[1], color='green', linestyle='--', label='95% CI Upper Bound')

plt.title("Normal Distribution of Simulated Data")
plt.xlabel("Data Values")
plt.ylabel("Density")
plt.legend()
plt.grid(alpha=0.3)
plt.show()
 
```

9. **Write a Python function to calculate the Z-scores from a dataset and
visualize the standardized data using a histogram. Explain what the Z-scores represent
in terms of standard deviations from the mean.**

    A **Z-score** tells us **how many standard deviations a data point is from the mean**.

    $$
    Z = \frac{(X - \mu)}{\sigma}
    $$

    - Z = 0 → value is **exactly at the mean**.  
    - Z = +1 → value is **1 standard deviation above the mean**.  
    - Z = -1 → value is **1 standard deviation below the mean**.  
    - Larger |Z| → more extreme the value relative to the mean.

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_zscores(data):
    # Calculate mean and standard deviation
    mean = np.mean(data)
    std = np.std(data, ddof=1)
    
    # Compute Z-scores
    z_scores = (data - mean) / std
    
    # Display summary
    print(f"Mean of Original Data: {mean:.2f}")
    print(f"Standard Deviation of Original Data: {std:.2f}")
    print(f"First 5 Z-scores: {z_scores[:5]}")
    
    # Plot histogram of Z-scores
    plt.figure(figsize=(8,5))
    plt.hist(z_scores, bins=10, color='lightcoral', edgecolor='black', density=True)
    plt.title("Histogram of Standardized Data (Z-scores)")
    plt.xlabel("Z-score")
    plt.ylabel("Density")
    plt.axvline(0, color='blue', linestyle='--', label='Mean (Z=0)')
    plt.legend()
    plt.grid(alpha=0.3)
    plt.show()
    
    return z_scores

# Example dataset
data = np.array([65, 70, 72, 68, 74, 77, 80, 85, 90, 95])

# Call function
z_scores = calculate_and_plot_zscores(data)
```
